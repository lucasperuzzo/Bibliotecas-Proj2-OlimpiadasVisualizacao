## 3. Brasil vs Mundo

Faça um gráfico de barras comparando os maiores medalhistas brasileiros com os maiores medalhistas do mundo em suas respectivas categorias.

Represente o esporte no eixo X, a quantidade de medalhas no eixo Y, coloque barras lado-a-lado representando os diferentes atletas de uma mesma modalidade e empilhe as medalhas de ouro, prata e bronze de cada atleta.

````python
lista_medalhistas_brasileiros_ordenados = df_historico_olimpiadas_br[~ df_historico_olimpiadas_br['Medal'].isna()].value_counts('Name')
lista_medalhistas_brasileiros_ordenados = lista_medalhistas_brasileiros_ordenados.head(5)
cinco_maiores_medalhistas_brasileiros = df_historico_olimpiadas_br[df_historico_olimpiadas_br['Name'].isin(lista_medalhistas_brasileiros_ordenados.index)]
````

````python
info_maiores_br = cinco_maiores_medalhistas_brasileiros.groupby(by=['Sport','Name',])['Medal'].count().reset_index()
medalhistas_mundo_modalidades_br = df_medalhistas_verao[df_medalhistas_verao['Sport'].isin(info_maiores_br['Sport'])]

medalhistas_mundo_modalidades_br = medalhistas_mundo_modalidades_br.groupby(by= ['Sport','Name'])['Medal'].count().sort_values(ascending=False)
esportes_comparados = medalhistas_mundo_modalidades_br.index.get_level_values(0).unique().to_list()
cinco_maiores_medalhistas_por_esporte = pd.DataFrame()
nomes_atletas = []

for esporte in esportes_comparados:
    nomes_atletas.extend(medalhistas_mundo_modalidades_br.loc[esporte].head(5).index.to_list())

nomes_atletas.extend(lista_medalhistas_brasileiros_ordenados.index.to_list())

atletas_comparados = df_medalhistas_verao[df_medalhistas_verao['Name'].isin(nomes_atletas)]
atletas_comparados = atletas_comparados.groupby(by=['Sport','Name','Medal'])['Season'].count().reset_index()
atletas_comparados = pd.pivot_table(atletas_comparados, index=['Sport','Name'], columns='Medal', values='Season').reset_index()
atletas_comparados.fillna(value=0, inplace=True)
````

<font color='blue'> Como a questão pede um gráfico de barras criei a função abaixo para criar esse gráfico. Agrupa-se os labels e empilha-se as categorias desejadas.  
    
Só não conseguir criar um label para cada grupo

````python
#Função para criar grupos de barras empilhadas. Autor: Lucas Peruzzo
import matplotlib.patches as mpatches

def grafico_agrupado_empilhado(df,agrupamento,label_barras,pilha,cores):

    dicionario_cores = pd.DataFrame(data=cores,index=pilha)

    qtdade_cat_empilhadas = len(pilha)
    df = df.groupby(by=[agrupamento,label_barras])[pilha].sum()
    grupos = df.index.get_level_values(0).unique()
    tamanho_eixo_x = len(grupos)

    largura = 0.8

    labels = []
    posicoes_labels = []

    tamanho_grupos = []
    dfs_grupos = []
    indices_dos_grupos = []
    
    for grupo in grupos:
        dfs_grupos.append(df.loc[grupo])
        indices_dos_grupos.append(df.loc[grupo].index)
        tamanho_grupos.append(len(df.loc[grupo]))
    
    posicao_barra = 0
    for x in range(tamanho_eixo_x):
        total_series = tamanho_grupos[x]
        posicao_barra += 1*largura + 0.05*largura
        for serie in range(total_series):
            ultima_categoria = None
            posicao_barra += 1*largura + 0.05*largura
            for categoria in pilha:

                if dfs_grupos[x].loc[indices_dos_grupos[x][serie]][categoria] != 0:  
                    labels.append(indices_dos_grupos[x][serie])
                    posicoes_labels.append(posicao_barra)

                    plt.bar(x=posicao_barra,height=dfs_grupos[x].loc[indices_dos_grupos[x][serie]][categoria], bottom=ultima_categoria, color=dicionario_cores.loc[categoria], width=largura);
                    if ultima_categoria == None:
                        ultima_categoria = dfs_grupos[x].loc[indices_dos_grupos[x][serie]][categoria]
                    else:
                        ultima_categoria += dfs_grupos[x].loc[indices_dos_grupos[x][serie]][categoria]
                        
    mapeamento_cores = []                
    for k in range(len(pilha)):
        mapeamento_cores.append(mpatches.Patch(color=cores[k], label=pilha[k]))
        
    plt.legend(handles=mapeamento_cores)
    plt.xticks(posicoes_labels, labels=labels,);
    
    return
````

````python
plt.figure(figsize=(16, 8))
plt.tick_params(labelsize=12)
plt.ylabel('QUANTIDADE DE MEDALHAS', fontsize=20)
plt.xlabel('ATLETAS AGRUPADOS POR ESPORTE', fontsize=20)
plt.grid(color='silver', linestyle='--', linewidth=1.2, axis='y', zorder=0);

grafico_agrupado_empilhado(atletas_comparados,'Sport','Name',['Gold','Silver','Bronze'],['gold', 'silver', 'darkorange'])
plt.xticks(rotation=80);
````
